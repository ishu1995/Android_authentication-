```python
# to convert number of ratings into sub groups
def rat_to_groups(no_of_ratings):
  '''
  function to convert ratings to resp translation
  '''
  if no_of_ratings <= 10:
    return 'ratings between 0-10'
  elif (no_of_ratings >= 10) and (no_of_ratings <1000 ):
    return 'ratings between 10 and 1000'
  else:
    return 'ratings greater than 1000'
```

```python
#plot for no of user ratings each app
df['Rating_cate_dummy'] = df['Number of ratings'].apply(lambda x: rat_to_groups(x) )
plt.figure(figsize=(15,9))
ax = sns.countplot(x="Rating_cate_dummy", data= df, palette='husl')
plt.xlabel('Sub groups ')
plt.ylabel('Number of applications')
plt.title('Number of users ratings each app')

ax1 = df.plot.scatter(x='Number of ratings', y = "Class") 
```

```python
# Fitting the model on the prepared dataset
df1=pd.read_csv('/content/drive/MyDrive/Almabetter/final_dataframe.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score, confusion_matrix, accuracy_score

from pylab import rcParams
rcParams['figure.figsize'] = 8, 8
```

```python
df1.head()
```

```python
df1.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
df1.head(2)
```

```python
# Fitting Random Forest on the dataset
X = df1.drop(labels=['Class'],axis=1)
```

```python
Y = df1['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0, stratify = Y)
```

```python
print("The shape of train dataset :")
print(X_train.shape)

print("\n The shape of test dataset :")
print(X_test.shape)
```

```python
print("Distribution of classes of dependent variable in train :")
print(Y_train.value_counts())

print("\n Distribution of classes of dependent variable in test :")
print(Y_test.value_counts())
```

```python
classifier = RandomForestClassifier() 
grid_values = {'n_estimators':[50, 80,  100], 'max_depth':[3, 5, 7]}
classifier = GridSearchCV(classifier, param_grid = grid_values, scoring = 'roc_auc', cv=5)
```

```python
classifier.fit(X_train, Y_train)
```

```python
train_preds =  classifier.predict(X_train)
test_preds  = classifier.predict(X_test)
```

```python
# Obtain accuracy on train set
accuracy_score(Y_train,train_preds)
```

```python
# Obtain accuracy on test set
accuracy_score(Y_test,test_preds)
```

```python
# Calculate roc_auc score on train set
roc_auc_score(Y_train,train_preds)
```

```python
# Calculate roc_auc score on test set
roc_auc_score(Y_test,test_preds)
```
